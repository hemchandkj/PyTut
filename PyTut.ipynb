# **ðŸ§­ Lesson 9: List Methods â€“ Adding, Removing & Slicing**

```python
# Playing around with lists
items = ['pen', 'pencil', 'eraser']
print(items)

# add item using .append()
items.append('scale')
print(items)

# insert sharpener at index 1
items.insert(1, 'sharpner')
print(items)

# extend the list
items.extend(['marker', 'highlighter'])
print(items)

# remove pencil
items.remove('pencil')
print(items)

# remove item using .pop() function. It removes an item from a specific index but .remove() function removes by name of the item
items.pop(2)
print(items)


# Slicing the lists
cities = ['A', 'B', 'C', 'D', 'E', 'F']
print(cities)

# print the first 3 cities
print(cities[:3])   # the last index, 3,  is excluded from the list

# print the last 2 cities
print(cities[-2:])

# print cities from index 2 to 4
print(cities[2:4])



# using del function
print(cities)

del cities[2:4 and 5] # beware: del is acted in the beginning of the list unlike other list functions that are acted eith . in the end with () or []
print(cities)
```

ðŸ§  Key Takeaways:

| Concept        | Method                 |
| -------------- | ---------------------- |
| Add to end     | `.append(item)`        |
| Insert at spot | `.insert(index, item)` |
| Add many       | `.extend([items])`     |
| Remove item    | `.remove(item)`        |
| Remove index   | `.pop(index)` / `del`  |
| Slicing        | `list[start:end]`      |


âœ… The last index is excluded from the list, list[start:end]

# **ðŸ§  Lesson 12: List Comprehension & Advanced Slicing**

Comprehension:  a NEW LIST by applying an EXPRESSION to each item in an existing list (or any iterable).

```python
# convert speed into cm/s

speeds_ms = [2.5, 5.0, 7.5, 10.0, 12.5]

speeds_cms = [100*v for v in speeds_ms]

print(speeds_cms)
print()

# extract the middle 4 temperatures

temps = [22, 24, 27, 30, 28, 26, 25, 29, 31, 23]

middle_temps = temps[3:-3]

print(middle_temps)
print()

# a new list for items names containg seven or fewer letters

inventory = ['pipette', 'beaker', 'test_tube', 'flask', 'burette', 'graduated_cylinder']


# Delete the variable named 'len' to unshadow the built-in function
# This variable was likely defined in a previous cell and is causing the error
if 'len' in globals() and isinstance(len, float):
    del len

short_names = [item for item in inventory if len(item)<=7]

print(short_names)

# the last 3 items in the inventory

new_inventory = inventory[-3:]
print(new_inventory)
```

* âœ… **List Comprehension Syntax**: `[expression for item in existing_list if condition]` creates a new list by transforming or filtering items in one line.
* âœ… **Basic Slicing**: `list[start:end]` returns items from index `start` (inclusive) to `end` (exclusive).
* âœ… **Advanced Slicing with Steps**: `list[start:end:step]` selects every `step`-th element between `start` and `end`.
* âœ… **Negative Indices in Slicing**: `-n` refers to the náµ—Ê° item from the end, so `list[-3:]` retrieves the last three elements.

ðŸ”‘ Key Takeaways: random Module

âœ… random.random()
Returns a float in the range [0.0, 1.0).

âœ… random.randint(a, b)
Returns an integer inclusive of both endpoints

âœ… random.choice(seq)
Selects a single, random element from any sequence (list, string, etc.).

âœ… random.shuffle(list)
Shuffles a list in place; does not return a value.

âœ… random.sample(seq, k)
Returns a new list of k unique elements drawn from the sequence.

âœ… random.seed(n)
Initializes the random number generator for reproducible results.

```python
# coin flip simulator: simulate 100 coin flips

import random

count_heads = 0
count_tails = 0

for i in range(100):
  flip = random.choice(['head', 'tail'])
  if flip == 'head':
    count_heads += 1
  else:
    count_tails += 1

print(f"Total heads: {count_heads}")
print(f"Total tails: {count_tails}")


print(f"Total counts: {count_tails + count_heads}")
```

```python
# random protein fragment
amino_acids = ['ALA','GLY','SER','VAL','LEU']

residues = random.sample(amino_acids, 3)

print(f"Three unique residues are: {residues}")

print("Selected residues:", ', '.join(residues)) # this is a new approach suggested
```

```python
# working with datetime module

from datetime import datetime, timedelta

# log entry timestamp
event = input("Please input an event description: ")
timestamp = datetime.now().strftime("[%Y-%m-%d %H:%M:%S]")
print(f"{timestamp} {event}")

```

```python

# parsing strings into dates

date_str = "13-06-2025"

date = datetime.strptime(date_str, "%d-%m-%Y")
print(date)
```

```python
# project deaadline calcualtor

start = input("Please input the start date: ")
days = int(input("Number of days: "))

start_dt = datetime.strptime(start, "%Y-%m-%d")

deadline_dt = start_dt + timedelta(days = days)

print("Deadline:", deadline_dt.strftime("%d/%m/%Y"))
```
